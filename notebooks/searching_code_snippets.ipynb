```python
def clean_text(text: str) -> str:
    text = re.sub(r"(:)\n{2}[\s\S]+?[\}\;]\n{3}", "\1 code ", text)  # code in spaces
    text = re.sub(r"\${1,2}[\s\S]+?\${1,2}", " equation ", text)  # latex equations
    return text
```
